```python
# Draw contour image

import matplotlib
import numpy as np
import matplotlib.cm as cm #cm is for color map
import matplotlib.pyplot as plt

delta = 0.025
x = np.arange(-3.0 ,3.0, delta)
y = np.arange(-2.0, 2.0, delta)
X, Y  = np.meshgrid(x,y)
Z1 = np.exp(-X**2 - Y**2)
Z2 = np.exp(- (X-1)**2 - (Y-1)**2)
Z = (Z1 - Z2) * 2

#print(X)
#print(Y) #Just to see the values, always a good practice when we are creating a complex data while studying
```

```python
fig , ax = plt.subplots()
CS = ax.contour(X,Y,Z) #X axis values, Y axis values and Z is height here
ax.clabel(CS, inline = 1, fontsize = 10)
ax.set_title('contour map') #You can imagine a data of hill and sea level or may profit and loss etc

#One can play around with these numbers to see the effect, I have not used it much but just taking this example to show youS
```

```python
# Draw a 3-D surface image

from mpl_toolkits.mplot3d.axes3d import Axes3D

fig = plt.figure(figsize = (14,6))

# Specify the 3D graphics to draw, with projection = '3d'

ax = fig.add_subplot(1, 2, 1, projection = '3d') #Lets's try different numbers for 1 ,2 and 1 and see the effect
ax.plot_surface(X, Y, Z, rstride=4, cstride = 4, linewidth=0)


#Now compare it with contour and try to analyse the two images
```

# Write a python program to plot a pie chart showing popularity of programming languages

```python
import matplotlib.pyplot as plt

# Data to plot

languages = 'Python' , 'Java', 'PHP', 'C#', 'C++'
popularity = [32.5, 10.5, 17, 20, 20]
colors = ['Blue', 'Green', 'Red', 'Yellow', 'Purple']
```

```python
# Explode 1st slice

explode = (0.1, 0, 0, 0, 0)
plt.pie(popularity)
```

```python
plt.pie(popularity, explode = explode, labels = languages, colors = colors, autopct = '%1.1f%%', shadow = True, startangle =0)
```
